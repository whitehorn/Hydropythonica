### Пакеты для научной работы

1. [Numpy](http://www.numpy.org/) - улучшенный модуль для математических вычислений;

2. [Scipy](http://scipy.org/) - пакет для научных вычислений, библиотека численного анализа;

3. [Matplotlib](http://www.matplotlib.org/) - графическая библиотека для получения графики анализа данных высокого качества;

4. [Basemap](http://matplotlib.org/basemap/) - модуль, поддерживающий работу с картографическими проекциями и shape-файлами;
*. Обучающие материалы (eng.) по работе с Basemap;

5. [Pandas](http://pandas.pydata.org/) - пакет для эффективной работы с таблицами и временными рядами;

6. [Seaborn](http://stanford.edu/~mwaskom/software/seaborn/tutorial/distributions.html) - пакет для быстрой статистической обработки и визуализации данных.
