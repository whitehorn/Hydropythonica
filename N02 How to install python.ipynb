## Установка python

В интернете достаточно легко найти материалы по установке python для различных ОС. Например, [вот такой](http://codeinlife.ru/python/python-nachalo-kak-ustanovit-python-v-windows-i-linux.html).

Большинство инструкций, особенно для новичков, рекомендует в качестве стандартного редактора кода IDLE, который поставляется со стандартной python-библиотекой.

Стандартная библиотека python для различных ОС доступна на [официальном сайте python](https://www.python.org/downloads/). Там же есть ссылка на [альтернативные реализации](https://www.python.org/download/alternatives/). Среди них есть замечательный для научной работы вариант - дистрибутив [Anaconda от Continuum Analytics](https://www.continuum.io/downloads).

Достоинств у этого варианта много:

+ кроссплатформенность;
+ доступны различные виды сборок: от минимальных до снабжённых практически всеми нужными пакетами;
+ удобный и мощный интерфейс поиска и установки дополнительных пакетов;
+ автодополнение кода;
+ удобный доступ к строкам документации.

Ниже приведена инструкция по установке Anaconda на Windows ОС.

### Загрузка дистрибутива

Скачиваем [python-дистрибутив Anaconda](https://www.continuum.io/downloads) от Continuum Analytics**

Перед установкой треуется пройти простую регистрацию. После указания электронной почты выбираем необходимую версию python. На момент написания (13.01.2017) в python 3.5 не поддерживается модуль basemap для windows-дистрибутива, необходимый для создания географических карт-схем.

Рекомендуется устанавливать актуальный выпуск 2-ой версии python (python 2.7 и выше).

Запускаем исполняемый файл и следеум инструкции по установке, оставляя значения по умолчанию.

Если всё прошло успешно, то python теперь установлен у вас на компьютере!

### Часто встречаемые проблемы

К сожалению, иногда установка происходит с ошибкой. Для ОС семейства Windows это чаще всего случается по следующей причине.

> В имени пользователя использана кириллица, т.е. русские буквы!!!

Если вы видите в командной строке что-то типа **C:\Users\Саша>**, то дело плохо. Под управлением такого пользователя дистрибутив Anaconda не установится и не будет работать. 

**Лечение**: необходимо создать нового пользователя с правами администратора в имени которого ТОЛЬКО ЛАТИНИЦА.

Пример хорошего имени пользователя: **C:\Users\sasha>**

В зависимости от версии Windows действия различны. [Инструкция для Windows 8.1 (об учётной записи с 1:13 по 1:58)](https://www.youtube.com/watch?v=Wq0fqnL8YGk) немного отличается от инструкции к [Windows 10](https://windowstips.ru/notes/18570).
 
### Добавление пакетов

В Anaconda уже установлены многие пакеты, необходимые для научной работы: numpy, scipy, matplotlib, pandas и др.

Если нужен какой-то пакет, а он не установлен, то его можно установить несколькими способами. 

Самый правильным является установка через conda-интерфейс. Он реализован через командную строку windows. В командной строке, которая вызывается командой cmd в окне "Выполнить" (комбинация клавиш Windows + R). В командной строке (из любого места ОС) вводим:

> conda install имя_пакета

Повторяем пункт 3 для следующих пакетов:

+ имя_пакета=basemap, то есть:

> conda install basemap
    
+ имя_пакета=netcdf4:

> conda install netcdf4

Numpy, scipy, pandas, sklearn и matplotlib уже входят в дистрибутив Anaconda.

### Редактор кода Spyder

Находим в ОС значок запуска ("иконку") редактора Spyder или в командной строке вводим:

> spyder.exe

![caption](files/pics/L2/spyder_window.png "Рисунок 1 Главное окно редактора Spyder")

Стандартно левая часть Spyder отдана под написание python-скрипта (скрпитовый режим), а правая нижняя часть - для консоли (интерактивный режим). Верхняя правая часть отдана под справочную информацию: 

+ Variable explorer - справочная таблица о свойствах переменных, хранящихся в данный момент в памяти;
+ Object Inspector - отображает справочную информацию при нажатии CTRL+I на наведённом участке кода (строка документации по выделенному элементу); 
+ File Explorer - навигация по ОС

Python-скрипт должен быть сохранён на жёстком или съёмном диске. Он имеет расширение ".py". Если скрипт сохранён, его можно запустить на исполнение через меню **"Run/Run"** или горячую клавишу **F5**. Такой способ равносилен запуску в консоли или ipython console команды **runfile('название_скрипта.py')** или **run('название_скрипта.py')** соответственно. Эти функции позволяют из консольного режима запустить на исполнение файл с python-кодом, как это делает интерпретатор из командной строки: **> python название_скрипта.py**

Работа в консоли или ipython console в целом аналогична работе со скриптом. Чтобы запустить на исполнение строку кода, нужно нажать **ENTER**. Если требуется ввести несколько строк кода, то их нужно набирать и исполнять последовательно. Причём при нажатии клавиши "Вверх" во время нахождения курсора в консоли перебираются последние набранные команды, что является весьма удобным инструментом при работе в консольном режиме. 

[Документация Spyder](https://pythonhosted.org/spyder/) описывает возможности пользовательской настройки этого свободно распространяемого редактора (IDE).

#### Подключаем pep8 

Стараемся с самого начала соблюдать стандарт написания и оформления кода - документ pep8. Чтобы Spyder подсказывал нам о несоответствиях между авторским кодом и pep8, необходимо зайти в "Tools/Preferences/Editor/CodeIntrospections//Analysis" и поставить галочку напротив style analysis (pep 8).

#### Приветствуем мир!

Набираем в консоли (после знака приглашения к вводу >>>)

> print('Hello, world!')

или в ipython console (см. "Tools/Preferences/Run/Console")

> In [1]: print('Hello, world!')

Что-то перестало работать? Не получается скачать нужный пакет или не проходит обновление? Переустановите anaconda - скачайте самую последнюю сборку и установите её предварительно удалив предыдущую версию.

## Интерактивный и скриптовый режим работы python

Существует два основных режима работы с python: **интерактивный** и **скриптовый**.

### Интерактивный режим

Интерактивный режим также ещё можно назвать консольным, так как он подразумевает работу в консоли python. При первом запуске стандартного редактора IDLE из базового дистрибутива python, открывается именно консоль. Её легко узнать по знаку приглашения в виде трёх стрелочек: >>>.

Такой режим подразумевает введение одной команду (или нескольких, отделяя каждую друг от друга точкой с запятой ";") и немедленное её выполнение интерпретатором python после нажатия клавиши Enter. Введя команду 

> print( 'Hello, world!')

и нажав клавишу Enter, на следующей строке будет отображён ответ: Hello, world!. Вывод команды print() осуществляется именно в консоль.

![caption](files/pics/L2/python_console_windows.png "Рисунок 2 Консольный режим python и ipython")

Но что делать, если нужно ввести много строк кода? А если будут ошибки? Неужели нужно будет вводить все предыдущие строки кода заново? Чтобы таких проблем у разработчиков и пользователей не было, в python есть скриптовый режим.

### Скриптовый режим

Скриптовый режим - это запуск текстового файла, сохранённого с расширением \textbf{".py"} или \textbf{".pyw"}, интерпретатором python.

#### Работа в командной строке windows

Для работы с python в консольном режиме из [командной строки windows (cmd)](https://ru.wikipedia.org/wiki/Cmd.exe) необходимо добавить путь к python в переменную окружения PATH. Дистрибутив Anaconda это делается автоматически (во время установки по умолчанию стоит специальная галочка). Для стандартного или иного дистрибутива в ОС Windows необходимо:

1. перейти в "Панель управления/Система и безопасность/Система/Дополнительные параметры в системе/Переменные среды";}
2. найти переменную PATH. Нажимаем "Изменить";}
3. в конце строки поставить ";" и набирать путь к интерпретатору python.exe (обычно это "C:/Python27").

После этого можно работать с python в консольном режиме и запускать python-скрипты с помощью командной строки windows из любого места на диске (т.е. вызывать интерпретатор через python.exe my_script.py или python script.py в командной строке). Пример запуска python-скрипта с именем **'myscript.py'** через командную строку выглядит так:

> **python myscript.py**

N.B. Для вызова скрипта вы должны находится в той же папке, где хранится файл скрипта (а это обычный текстовый файл с расширением "py"). Либо вы должны указывать полный путь к файлу-скрипту, как показано ниже. Так, если скрипт называется **'script1.py'** и он расположен по адресу "c:\Users\Ivan\pythonLib\", то полный путь 

> **python c:\Users\Ivan\pythonLib\script1.py**

![caption](files/pics/L2/test_script.png "Рисунок 3 Запуск скрипта и вывод в консоль cmd")

Работа в командной строке позволяет лучше понять работу python в ОС семейства Linux. Набрав в командной строке просто "python", появится приглашение интерпретатора python: >>>.

> Чтобы **выйти из интерактивного режима** python обратно в командную строку, нужно ввести команду **quit()** или нажать сочетание клавиш **CTRL+D**.
