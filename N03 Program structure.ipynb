## Программа и её структура

[Программа](https://ru.wikipedia.org/wiki/%D0%9A%D0%BE%D0%BC%D0%BF%D1%8C%D1%8E%D1%82%D0%B5%D1%80%D0%BD%D0%B0%D1%8F_%D0%BF%D1%80%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B0) - это последовательность инструкций, определяющих процедуру решения конкретной задачи. В общем случае инструкции выполняются одна за одной, последовательно. Состав инструкций и их последовательность определяет пользователь-программист.

С точки зрения программиста программа на конкретном языке - это [алгоритм](https://ru.wikipedia.org/wiki/%D0%90%D0%BB%D0%B3%D0%BE%D1%80%D0%B8%D1%82%D0%BC) и [структура данных](https://ru.wikipedia.org/wiki/%D0%A1%D1%82%D1%80%D1%83%D0%BA%D1%82%D1%83%D1%80%D0%B0_%D0%B4%D0%B0%D0%BD%D0%BD%D1%8B%D1%85). Алгоритм отвечает на вопрос "что?", а структура данных - "как?". Освоение нового языка программирования по сути означает изучение лишь структуры данных, а алгоритмирование остаётся в стороне, являясь теоретической базой любого программиста.

Алгоритм всегда можно записать в виде [блок-схемы](https://ru.wikipedia.org/wiki/%D0%91%D0%BB%D0%BE%D0%BA-%D1%81%D1%85%D0%B5%D0%BC%D0%B0), который не зависит от конкретного языка программирования.

### Стили программирования

Вид кода во многом определяется используемой парадигмой программирования. Python является мультипарадигмальным языком, то есть он поддерживает несколько парадигм программирования:

+ [процедурный](https://ru.wikipedia.org/wiki/%D0%9F%D1%80%D0%BE%D1%86%D0%B5%D0%B4%D1%83%D1%80%D0%BD%D0%BE%D0%B5_%D0%BF%D1%80%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5);

+ [объектно-ориентированный](https://ru.wikipedia.org/wiki/%D0%9E%D0%B1%D1%8A%D0%B5%D0%BA%D1%82%D0%BD%D0%BE-%D0%BE%D1%80%D0%B8%D0%B5%D0%BD%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%BD%D0%BE%D0%B5_%D0%BF%D1%80%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5);

+ [функциональный](https://ru.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D0%BE%D0%BD%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D0%BF%D1%80%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5);

Можно придерживаться как одного стиля, так и смешивать стили при кодировании: чётких границ и ограничений здесь нет. Наоборот, свобода в использовании разных подходов к описанию алгоритма является очень ценным преимуществом языка python. 

Самым распространённым является **процедурный стиль**. Он подразумевает, что последовательность команд-операторов можно собрать в более крупные, целостные единицы кода с помощью механизмов самого язык, то есть разбить программу на подпрограммы.

Само название (**ОБЪЕКТНО-ОРИЕНТИРОВАННЫЙ стиль** или программирование, ООП) говорит о том, что этот стиль ориентируется на работу с некоторым объектом. Проще всего представить реальный объект и описать его некоторыми параметрами. Пример: магазин, банк, школа, спортсмен, клиент, пациент. Более абстрактный пример объекта - поле для игры в шахматы, тетрадь. В основе ООП лежит понятие класса или объекта как абстрактной программной единицы. В python их принято называть классами. В python даже типы данных (строки, числа и др.) также явлются классами.

**Функциональный стиль** исходит из парадигмы, противоположной императивной. Как следует из названия, в данном стиле акцент сделан на рассмотрение функций. Так процесс вычисления здесь трактуется как вычисление значений функций в математическом понимании последних (в отличие от функций как подпрограмм в процедурном программировании). Функциональный стиль частично [поддерживаестя в python](https://ru.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D0%BE%D0%BD%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D0%BF%D1%80%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5_%D0%BD%D0%B0_Python) и выражается в хорошей поддержке рекурсии, функций высших порядков (функции могут принимать в качестве входящих параметров функции), декораторов и др.

## Составные части программы на python

Python-программа как файл-объект - это текстовый файл с расширением ".py". Он может называться:

+ python-программа;

+ python-скрипт;

+ python-модуль.

Структурно можно выделить следующие составные части python-программы:

+ главный модуль (обязательно);

+ модуль (необязательно);

+ класс (характерно для ООП стиля; необязательно);

+ функция (необязательно).

Существует руководящий документ для оформления кода на python (он же [pep8](http://pythonworld.ru/osnovy/pep-8-rukovodstvo-po-napisaniyu-koda-na-python.html)). [Кратко эти правила ](http://rukeba.com/by-the-way/pep8-korotko-i-po-russki/
) регламентируют как именно нужно кодировать те или иные части программы.

Там рекомендуется, например, осуществлять импорт модулей в начале любого модуля-программы. Стоит придерживаться следующего порядка импортирования: стандартные модули (os, time, sys и др.), сторонние модули (numpy, pandas, netcdf4 и др.), локальные или пользовательские модули (my_module, например). Лучше при этом отделять группы импорта пустой строкой.

**Следование pep8 - это хороший тон и поддержка высокого уровня сопровождения кода.**

### Модуль

Python является [модульным языком](https://ru.wikipedia.org/wiki/%D0%9C%D0%BE%D0%B4%D1%83%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D0%BF%D1%80%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5) программирования.

Python-программа всегда имеет один главный модуль. Вне зависимости от названия скрипта-файла, его внутреннее имя будет '__main__'. Помимо главного модуля, составяющего ядро программы, в программу могут входить и другие модули или просто модули. Они выполняют очень важную роль: предоставляют доступ к вложенному в них коду, расщиряют функциональность главного модуля.

Все модули логически равнозначны, но только один модуль является главным и составляет "ядро программы".

### Функция

Подрограмма, как программная единица, в python реализована в виде **функции def()**. Внутри функции может располагаться последовательность команд-операторов, которая часто встречается в программе. Функции могут быть описаны как в том же файле, что и главный модуль. А могут хранится и в другом, стороннем модуле и импортироваться в главный.

## Управление потоком выполнения программы

Вне зависимости от способа организации кода (работа с классами, использование модулей и функций) решать практические задачи с помощью только последовательного исполнения строк кода практически невозможно. Почти всегда требуется возможность изменить ход выполения код, изменить состав выполняемых инструкций в зависимости от каких-либо условий. 

Для управления потоком выполняемых команд для произвольной совокупности строк кода (в том числе и внутри функций) служат особые конструкции-операторы - управляющие блоки. К блокам относятся:

+ **циклы**;

+ **ветвление** (логические вентили; условные операторы);

+ **обработчики исключений**.

Блоки ограничивают некоторый участок кода, последовательность строк кода, заключая их в себя. Блоки могут содержать внутри себя строки кода, а также другие блоки, т.е. управляющие блоки могут быть вложенными.

Циклы позволяют повторять некоторый участок кода (объединение строк кода, лежащих внутри блока или тела цикла) или выполнение какого-либо оператора (в том числе и вызов функции). 

Ветвление же позволяет менять ход выполнения последовательности кода исходя из решений некоторого условия или ряда условий.

Обработчики исключений перехватывают ошибки, возникающие по ходу выполнения программы, и позволяют работать алгоритму стабильно, предлагая различные варианты обхода возникшего затруднения.

Циклы и ветвление - это базовые понятия, существующие практически во всех языках программирования.

#### Анекдот про условия и циклы

Жена посылает программиста в магазин: 
- Дорогой, купи, пожалуйста, палку колбасы, и если будут яйца, то купи десяток. 
Через полчаса программист возвращается с десятью палками колбасы. 
Жена: 
- Что это?! Зачем ты купил столько колбасы? 
Программист: 
- Ну так яйца-то были...

Соль этого анекдота легко представить в виде следующего кода.

```python
# Пример-анекдот глазами мужа-программиста

eggs = True   
bag = []   # сумка покупателя-программиста

for i in range(10):
    if eggs:   # проверка на наличие в магазине яиц
        bag.append('snag')

print 'Total programmer basket'
for i, item in enumerate(bag):
    print i+1, item
```

```python
# Пример-анекдот глазами жены

eggs = True   
bag = []   # сумка покупателя-программиста

bag.append('snag')
if eggs:
    for i in range(10):
        bag.append('egg')
        
print 'Total wife basket'
for i, item in enumerate(bag):
    print i+1, item
```

## Практика

Главный модуль - это исходное пространство, которое можно дополнить строками кода, блоками кода, оформленного в виде функций, а также кодом из других модулей.

Модули импортируются с помощью функции **import**. Чтобы импортироват модуль необходимо знать его имя. В python имеются стандартные модули, которые поставляются прямо с python (т.н. "батарейки").

```python
# Некоторые стандартные модули: sys, os, time, datetime

import os, sys   # импорт 3 стандартных модулей
import datetime  

print(sys.prefix)   # папка установки интерпретатора python

print(os.getcwd())   # адрес текущей папки

print(datetime.datetime.now())   # текущая дата и время 
```

Примеры описаний функций и классов. Описание или декларирование функций/классов не означает исполнения заключенных в них строк кода. Для их исполнения необходимо вызывать функцию или создать экземпляр класса.

```python
import os   # импорт модуля os для работы с операционной системой

def simple_function():
    print('Hello, World!')

class My_class():
    '''
    строка документирования. Она же комментарий к классу
    '''
    # Функция, принадлежащая классу, называется методом
    def f_empty():   # метод f_empty
        pass   # пустая команда, позволяющая существовать функции без других строк кода 
        
    print('Welcome to class!')

    simple_function()   # вызов функции из тела класса. Функция не принадлежит классу

# Для исполнения кода класса, необходимо создать его экземпляр

'''
Начало главной, исполняемой части программы
''' 

c = My_class()   # создание экземпляра класса. Экземпляром является переменная "c"
print(os.name)   # печать в консоль имени ОС
```

Приведённый выше пример также иллюстрирует структуру python-программы. Сам код - это главный модуль, в котором:

1. импортирован модуль os;

2. описана функция simple_function;

3. описан класс My_class;

4. создан экземпляр класса My_class, который вызывает функцию simple_function;

5. Выводится в консоль название операционной системы с помощью модуля os

### Итоги

Python поддерживает несколько парадигм программирования и реализует смешанную поддержку одноимённых стилей программирования. Базовым для изучения является процедурный стиль, основными понятиями которого являются подпрограмма или функция. Объектно-ориентированный стиль оперирует понятием класса, как абстрактного объекта, требующего описания и функционирования. Функциональный стиль реализован частично и выражается в виде поддержки рекурсии, вызова функциями функций и др. Все стили могут использоваться одновременно. Деление на стили условное и является относительной характеристикой организации кода.

Программа на python состоит хотя бы из одного текстового файла с расширением ".py" или ".pyw", содержащего python-код. 

Python-код может быть структурно разбит на составляющие части - функции, то есть представлен в более генерализованном виде. В предельном случае программу можно представить в виде последовательности команд, вызывающих функции. 

Характер и направление исполнения команд-операторов, выражений и вызова функций можно регулировать с помощью управляющих блоков - циклов, ветвления и обработчиков исключений. 

Циклы позволяют повторять заключенные внутри тела цикла строки кода, а ветвление обеспечивает контроль над ходом течения программы. Обработчики исключений позволяют программе продолжить исполнение даже при возникновении ошибки.

Свойство модульности позволяет логичнее и прозрачнее организовывать и хранить код, а также заметно облегчает использование программных единиц с едиными именами. Все модули могут содержать описания функций, классов (ООП) и строки кода, а также вызывать функции и содержание других модулей. При этом главным может быть только один модуль.
