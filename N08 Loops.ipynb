## Циклы

Одним из базовых понятий программирования (в процедурном стиле и не только) является цикл. [Цикл](https://ru.wikipedia.org/wiki/%D0%A6%D0%B8%D0%BA%D0%BB_%28%D0%BF%D1%80%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5%29) - это инструмент, позволяющий повторять какой-либо кусок кода много-много раз. Разумно использовать цикл столько раз, сколько это нужно для реализации алгоритма.

Каждый повтор называется ещё итерацией.

**Цикл (loop)** - это логический блок, конструкция-контейнер, в пределах которого осуществляется повторение всего кода, содержащегося внутри него. Всё лежащее внутри цикла отделяется от строки, в которой объявлен цикл, четырьмя пробелами (правило 4 пробелов для вложенных структур). Строка, в которой объявлен цикл, обязательно заканчивается двоеточием!

В python реализовано два вида циклов: конечные (for) и бесконечные (while). В первом случае известно количество повторений, шагов цикла или итераций. Во втором случае цикл будет повторять код бесконечно, если не будет выполнено условие выхода или не будет прервано пользователем вручную (это приведёт к остановке выполнения программы).

## Цикл for

Цикл for имеет следующую конструкцию:

**for <объект цикла> in <последовательность>**: 

**....тело цикла**

Уже знакомый оператор вхождения **in** здесь работает в связке с for и позволяет перебрать элемента последовательности. В конце условия цикла обязательно ставится двоеточие. Тело цикло должно быть вложенным, т.е. размещаться на 4 пробела правее от начала описания цикла (от позиции буквы "f" в for).

> Обратите внимание, что даже в интерактивном режиме (консоль) необходимо делать 4 пробела при написании кода внутри цикла! IDLE и другие работающие с python среды будут вам помогать делать это автоматически, но за правильностью отступов вложенных логических структур в коде нужно следить!

Важно помнить, что тело цикло должно быть! Пустая строка не годится, нужен какой-либо код. Чтобы описать "нулевой" или пустой цикл, который будет корректно восприниматься интерпретатором python, нужно воспользоваться особой командой **pass**. Это  пустая функция, пустая операция или выражение, которое позволяет управляющим структурам (циклы и ветвление) иметь тело без дополнительного кода. Это удобно в случае разработки кода, когда точно есть потребность в конструкции, но конкретное наполнение пока не определено. 

```python
a = range(10)   # список 
b = tuple(a)   # кортеж; функция tuple() возвращает кортеж от последовательности

# Перебор элементов списка b
for j in b:
    # j является элементом списка b
    print j,   # запятая подавляет перевод строки
# Перебор элементов кортежа a
for i in a:
    print i
```

В случае *словарей*, у которых элемент последовательности состоит не только из значения, а из связки **"ключ:значение"**, цикл for будет перебирать **ключи**. Чтобы узнать значения, соответствующие этим ключам, необходимо обратиться к словарю по ключу, используя квадратные скобки [].

Так как словарь является неупорядоченной последовательностью, т.е. в нём нет строго правила следования элементов друг за другом, то и порядок возвращаемых ключей/значений не соответствует порядку их ввода при составлении словаря.   

```python
c = {'one':1, 'two':2, 'three': 'long'}

# Перебор ключей, значений словаря c
print 'I'
for i in c:
    print i, c[i]
    
# Другой вариант
print 'II'
for key in c.keys():
    print key, c[key]
```

Одной из самых полезных функций при работе с циклами является функция **enumerate()**. Она применяется к последовательности и позволяет перебрать не только сами элементы последовательности, но и их индексы. Для этого указывается не одна переменная цикла (i в примере выше), а две, указанные через запятую: index, evalue.

> **enumerate()** - функция, которая возвращает на каждом шаге цикла индекс и значение элемента последовательности, которая перебирается в цикле

```python
c = ['windows', 'linux', 'mac', 'sun', 'unix']
for i in c:
    print(i)

for i, elem in enumerate(c):
        print(i, elem)
```

Как видно, цикл for позволяет работать не только с одной переменной (вспомним, что она есть ссылка на объект), а с несколькими переменными на каждом шаге (итерации) цикла. Поясним это на примере вложенных списков.

```python
a = ['Tom', 'Jerry', 'DOG']
b = range(3)
c = (True, False, True)
ilist = [a, b, c]   # вложенный список: список, состящий из списков
for i, j, k in ilist:   # здесь i - число, строка или булева переменная
    print type(i), i,j,k

print '*'*10
    
for i in ilist:   # здесь i - список
    print type(i), i
```

## Бесконечный цикл или цикл с условием while

Бесконечный цикл **while** в python имеет следующую конструкцию:

while(<выражение>):

....тело цикла

На каждом шаге цикла выражение, стоящее в скобках, проверяется на равенство истине (True). Если оно истинно, то совершается ещё один шаг, ещё одна итерация цикла. Бесконечным такой цикл делает выражение, которое всегда соответствует истине True на любом шаге такого цикла.

Обычно в качестве выражения в такой цикл помещают логическое условие: например, a > 10 (a > b; i < 10; d == 10). Но это может быть и просто переменная (ведь любой тип данных можно привести к логическому типу и сравнить на равенство Истине True). 

Соответственно, чтобы цикл не превратился в бесконечный, выражение цикла не должно быть константой, а некоторой переменной, управляемой внутри цикла. Так в случае выражения-условия a > 10 в теле цикла (участок кода, который охватывает цикл) значение переменной "a" должно изменяться и при определённых обстоятельствах стать меньше 10. Тогда такой код будет корректен, цикл не будет выполняться бесконечно, а интерпретатор покинет цикл в момент прекращения выполнения условия. Поэтому цикл **while** также называют **циклом с условием**.

```python
a = 10
while a > 5:
    print(a)
    a = a - 1
```

### Прерывание цикла break и продолжение итераций цикла continue

Чтобы выйти из цикла в определённый момент (даже если выражение цикла while истинно) существует оператор **break**. Как только он выполняется, программа покидает вышестоящий цикл и переходит к выполнению следующих за этим циклом строк кода.

В противоположность функции **break**, есть функция продолжения цикла, перехода на следующую итерацию - функция **continue**. Она позволяет пропустить оставшееся тело цикла, то есть код, который расположен ниже функции continue, и перейти к следующему шагу цикла.

Хотя такие конструкции и затрудняют разбор кода, они полезны, так как помогают прекратить выполнение ненужных участков тела цикла или итераций. **Break** и **continue** чаще всего встречаются в блоках условных операторов (if-elif-else).

```python
import random   # импорт модуля для работы со случайными числами


TR = 0.7
t = 0
while(True):
    t = t + 1
    a = a - 1
    c = random.random()   # генерация псевдослучайного числа
    if c >= TR:
        print(u'I. Succsess after', t, 'iterations')
        break

'''
В случае строго описания условий (два исхода - больше или равно \
некоторого числа или меньше его), вторая ветвь должна, как и всякий \
контейнер, содержать хоть какую-нибудь строку кода. В данном случае \
отлично подходит оператор continue.
'''
        
TR = 0.95
t = 0
while(True):
    t = t + 1
    a = a - 1
    c = random.random()   # генерация псевдослучайного числа
    if c >= TR:
        print(u'II. Succsess after', t, 'iterations')
        break
    else:   
        continue
```

## Задача о Гомере Симпсоне и пончиках

Приведём задачу про [Гомера Симпсона](https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D0%BC%D0%B5%D1%80_%D0%A1%D0%B8%D0%BC%D0%BF%D1%81%D0%BE%D0%BD) и пончики. Пара десятилетия назад аналогичная задача формулировалась относительно Карлсона и плюшек.
Задача: Гомер купил P пончиков и начал их есть. Доктор Хьюберт разрешает Гомеру съедать в день только NMAX пончиков (P > NMAX). Требуется:

+ вывести на экран количество съеденных Гомером пончиков;

+ алгоритм должен использовать бесконечный цикл while;

+ напишите алгоритм так, чтобы Гомер съел только разрешённое число пончиков.

![caption](files/pics/L7/homer_simpson.jpg "Рисунок 1 Гомер Симпсон ест пончики и не может остановиться")

```python
P = 100    # количество пончиков
NMAX = 20   # столько может съесть Гомер Симпсон и не попасть в больницу
N = 0    # столько пончиков Гомер уже съел
while NMAX:   # бесконечный цикл
    NMAX = NMAX - 1
    N = N + 1

print 'Гомер съел', N, 'пончиков'
```

Можно переписать код следующим образом, задействовав оператор **break**.

```python
P = 100    # количество пончиков
NMAX = 20   # столько может съесть Гомер Симпсон и не попасть в больницу
N = 0   # столько пончиков Гомер уже съел
while(P):   # бесконечный цикл
    P = P - 1
    N = N + 1
    if N >= NMAX:   # условие остановки цикла
        print('Homer, stop eating donuts!')
        break   # выход из цикла

print 'Гомер съел', N, 'пончиков'
```

В таком варианте число P является условием повторения цикла. Цикл закончится, когда P станет равным нулю (целый нуль соответствует логическому False). Но в этом случае Гомер съест пончиков больше, чем ему разрешает доктор (число NMAX). Поэтому для сохранения здоровья Гомера, ставится условие на досрочное завершение цикла.
